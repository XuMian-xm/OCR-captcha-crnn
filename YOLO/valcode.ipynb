```python
# Êï∞ÊçÆÈõÜÂàíÂàÜ
import os, shutil
from sklearn.model_selection import train_test_split


val_size = 0.2
#test_size = 0.2
postfix = 'png'
imgpath = r'./data/valcode/images'
txtpath =  r'./data/valcode/labels'

output_train_img_folder =r'./data/valcode/images/train'
output_val_img_folder =  r'./data/valcode/images/val'
output_train_txt_folder =  r'./data/valcode/labels/train'
output_val_txt_folder =  r'./data/valcode/labels/val'

os.makedirs(output_train_img_folder, exist_ok=True)
os.makedirs(output_val_img_folder, exist_ok=True)
os.makedirs(output_train_txt_folder, exist_ok=True)
os.makedirs(output_val_txt_folder, exist_ok=True)


listdir = [i for i in os.listdir(txtpath) if 'txt' in i]
train, val = train_test_split(listdir, test_size=val_size, shuffle=True, random_state=0)

#todoÔºöÈúÄË¶ÅtestÊîæÂºÄ

# train, test = train_test_split(listdir, test_size=test_size, shuffle=True, random_state=0)
# train, val = train_test_split(train, test_size=val_size, shuffle=True, random_state=0)

for i in train:
    img_source_path = os.path.join(imgpath, '{}.{}'.format(i[:-4], postfix))
    txt_source_path = os.path.join(txtpath, i)

    img_destination_path = os.path.join(output_train_img_folder, '{}.{}'.format(i[:-4], postfix))
    txt_destination_path = os.path.join(output_train_txt_folder, i)

    shutil.copy(img_source_path, img_destination_path)
    shutil.copy(txt_source_path, txt_destination_path)

for i in val:
    img_source_path = os.path.join(imgpath, '{}.{}'.format(i[:-4], postfix))
    txt_source_path = os.path.join(txtpath, i)

    img_destination_path = os.path.join(output_val_img_folder, '{}.{}'.format(i[:-4], postfix))
    txt_destination_path = os.path.join(output_val_txt_folder, i)

    shutil.copy(img_source_path, img_destination_path)
    shutil.copy(txt_source_path, txt_destination_path)

```

```python
from ultralytics import YOLO
model = YOLO("./runs//train/exp12/weights/best.pt")  
# Run batched inference on a list of images

char_to_idx = {
    'a': 0, 'b': 1, 'c': 2, 'd': 3, 'e': 4, 'f': 5, 'g': 6, 'h': 7, 'i': 8, 'j': 9,
    'k': 10, 'l': 11, 'm': 12, 'n': 13, 'o': 14, 'p': 15, 'q': 16, 'r': 17, 's': 18,
    't': 19, 'u': 20, 'v': 21, 'w': 22, 'x': 23, 'y': 24, 'z': 25, 'A': 26, 'B': 27,
    'C': 28, 'D': 29, 'E': 30, 'F': 31, 'G': 32, 'H': 33, 'I': 34, 'J': 35, 'K': 36,
    'L': 37, 'M': 38, 'N': 39, 'O': 40, 'P': 41, 'Q': 42, 'R': 43, 'S': 44, 'T': 45,
    'U': 46, 'V': 47, 'W': 48, 'X': 49, 'Y': 50, 'Z': 51, '0': 52, '1': 53, '2': 54,
    '3': 55, '4': 56, '5': 57, '6': 58, '7': 59, '8': 60, '9': 61
}

# Â¶ÇÊûúÈúÄË¶ÅÂèçÂêëÊò†Â∞ÑÔºà‰ªéÁ¥¢ÂºïÂà∞Â≠óÁ¨¶Ôºâ
idx_to_char = {idx: char for char, idx in char_to_idx.items()}

def test(path):
    results = model([path], verbose=False)
    fin = ''
    
    for result in results:
        xywh = result.boxes.xywh  # center-x, center-y, width, height
        names = [result.names[cls.item()] for cls in result.boxes.cls.int()]  # class name of each box
        confs = result.boxes.conf  # Ëé∑ÂèñÁΩÆ‰ø°Â∫¶‰ø°ÊÅØ
        
        # ÂàõÂª∫ÂåÖÂê´xÂùêÊ†á„ÄÅÁ±ªÂà´ÂíåÁΩÆ‰ø°Â∫¶ÁöÑÂàóË°®
        detections = []
        for index in range(len(xywh)):
            detections.append([
                xywh[index][0].item(),  # xÂùêÊ†á
                names[index],           # Á±ªÂà´ÂêçÁß∞
                confs[index].item()     # ÁΩÆ‰ø°Â∫¶
            ])
        
        # Â¶ÇÊûúÊ£ÄÊµãÂà∞ÁöÑÂ≠óÁ¨¶Êï∞ÈáèÂ§ß‰∫é6‰∏™ÔºåÊåâÁΩÆ‰ø°Â∫¶ÊéíÂ∫èÂπ∂ÂèñÂâç6‰∏™
        if len(detections) > 6:
            # ÊåâÁΩÆ‰ø°Â∫¶‰ªéÈ´òÂà∞‰ΩéÊéíÂ∫è
            detections.sort(key=lambda x: x[2], reverse=True)
            # Âè™‰øùÁïôÂâç6‰∏™
            detections = detections[:6]
        
        # ÊåâxÂùêÊ†á‰ªéÂ∑¶Âà∞Âè≥ÊéíÂ∫è
        detections.sort(key=lambda x: x[0])
        
        # ËΩ¨Êç¢Âπ∂ÊãºÊé•Â≠óÁ¨¶
        for det in detections:
            na = idx_to_char[int(det[1])]
            fin = str(fin) + na
            
    return fin
print(test(f'./data/valcode/images/train/05Faiu_895.png'))
```

```python
import os
import glob

def get_image_labels(directory_path):
    """
    ËØªÂèñÊåáÂÆöË∑ØÂæÑ‰∏ãÁöÑÊâÄÊúâPNGÊñá‰ª∂ÔºåÊèêÂèñÂÖ∂Êñá‰ª∂ÂêçÂâçÁºÄ‰Ωú‰∏∫Ê†áÁ≠æ
    
    ÂèÇÊï∞:
        directory_path: ÂåÖÂê´PNGÊñá‰ª∂ÁöÑÁõÆÂΩïË∑ØÂæÑ
        
    ËøîÂõû:
        ‰∏Ä‰∏™ÂàóË°®ÔºåÊØè‰∏™ÂÖÉÁ¥†ÊòØ‰∏Ä‰∏™ÂÖÉÁªÑ (Êñá‰ª∂Ë∑ØÂæÑ, Ê†áÁ≠æ)
    """
    result = []
    
    # Ëé∑ÂèñÁõÆÂΩï‰∏ãÊâÄÊúâÁöÑPNGÊñá‰ª∂
    png_files = glob.glob(os.path.join(directory_path, "*.png"))
    
    for file_path in png_files:
        # Ëé∑ÂèñÊñá‰ª∂ÂêçÔºà‰∏çÂê´Ë∑ØÂæÑÔºâ
        file_name = os.path.basename(file_path)
        
        # ‰ªéÊñá‰ª∂Âêç‰∏≠ÊèêÂèñÊ†áÁ≠æÔºà‰∏ãÂàíÁ∫øÂâçÁöÑÈÉ®ÂàÜÔºâ
        if '_' in file_name:
            label = file_name.split('_')[0]
            result.append((file_path, label))
        else:
            print(f"Ë≠¶Âëä: Êñá‰ª∂Âêç {file_name} ‰∏çÂåÖÂê´‰∏ãÂàíÁ∫øÔºåÊó†Ê≥ïÊèêÂèñÊ†áÁ≠æ")
    
    return result


# ÊåáÂÆöÁõÆÂΩïË∑ØÂæÑ
directory_path = "./data/valcode//images/train"

# Ëé∑ÂèñÊñá‰ª∂ÂíåÊ†áÁ≠æ
image_labels = get_image_labels(directory_path)

# ÁªüËÆ°‰∏çÂêåÊ†áÁ≠æÁöÑÊï∞Èáè
label_counts = {}
for _, label in image_labels:
    if label in label_counts:
        label_counts[label] += 1
    else:
        label_counts[label] = 1

total = 0
right = 0

for _ in image_labels:
    path = _[0]
    label = _[1]
    predict = test(path)
    print("label: {}, predict: {}".format(label, predict))
    total += 1
    if label == predict:
        right += 1

print("acc: {}".format(right/total))
    
```
